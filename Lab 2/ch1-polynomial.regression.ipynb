```python
import numpy as np

Time = np.array([6, 8, 11, 14, 16, 18, 19])
Temp = np.array([4, 7, 10, 12, 11.5, 9, 7])

import matplotlib.pyplot as plt
#we will show the temperature at a few points during the day:
plt.figure()
plt.plot(Time, Temp, 'bo')
plt.xlabel("Time")
plt.ylabel("Temp")
plt.title('Temperature versus time')
plt.show

"""
1. numpy.polyfit() function returns the coefficients for 
a polynomial of degree n (given by us) that is the best fit for the data.
2. The coefficients returned by the function are in descending powers 
(highest power first), and their length is n+1 if n is the degree of the polynomial.
"""
beta = np.polyfit(Time, Temp, 4)

"""
1. verify that it actually fits our data.
2. use the model to evaluate the polynomial at uniformly spaced times
3. use poly1d function
4. This function returns the value of a polynomial of degree n evaluated 
at the points provided by us
5. The input argument is a vector of length n+1 whose elements are 
the coefficients in descending powers of the polynomial to be evaluated:
"""
p = np.poly1d(beta)

# plot the original data and the model on the same plot
# In regression analysis, it's important to keep the order of the model as low as possible.
# linespace -> Return evenly spaced numbers over a specified interval.
xp = np.linspace(6, 19, 100)
plt.figure()
## plot x and y using blue circle markers
plt.plot(Time, Temp, 'bo', xp, p(xp), '-')
plt.xlabel("Time")
plt.ylabel("Temp")
plt.title('Temperature versus time')
plt.show
```
